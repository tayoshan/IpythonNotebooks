```python
import pandas as pd
import numpy as np
import math
from scipy.stats.stats import pearsonr
data = pd.DataFrame(pd.read_csv("C:\\Users\\Taylor\\Documents\\SIMODELTESTDATA2.csv"))
origins = 'Origin'
destinations = 'Destination'
sep = 'Dij'
tot_ratio = 1

pop = 'Pop'
#data = data[data[origins] !=data[destinations]]
```

```python
def s(data, origins, destinations, sep, pop, i, j):
    subset = data[(data[origins] == i)]
    sVals.append(subset[((subset[sep] < subset[sep][((subset[origins] == i) & (subset[destinations] == j))].values[0]) & (data[origins] !=data[destinations]))]['Pop'].sum())
```

```python
def calcFlows(data, x, tot_ratio):
    mi = data['Pop'][data['Destination'] == x.Origin].values[0]
    Ti = np.sum(data.Data)*tot_ratio
    flows.append((float(Ti)*float(mi)*float(x.Pop))/((mi + x.s)*(mi + x.Pop + x.s)))
```

```python
sVals = []
data.apply(lambda x: s(data, origins, destinations, sep, pop, x.Origin, x.Destination), axis=1)
data['s'] = sVals
flows = []
data.apply(lambda x: calcFlows(data, x, tot_ratio), axis=1)
data['SIM_Estimates'] = flows
```
