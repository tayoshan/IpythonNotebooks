```python
import geopandas as gp
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import itertools as it
import time
%pylab inline
```

```python
london = gp.read_file('C:/londonSHP/London_Ward_CityMerged.shp')
stations = gp.GeoDataFrame(gp.read_file('C:/Users\Taylor/Desktop/barclayscyclehireusagestats2012/barclayscyclehireusagestats2012/uniqueStations.shp'))
stations.crs = {'init': 'epsg:4326', 'no_defs': True}
stations.to_crs(crs=london.crs, inplace=True)
```

```python
ward_with_stations = london[london.geometry.intersects(stations.geometry.unary_union) == True]
ward_with_stations.plot()
```

```python
unique = ward_with_stations.NAME.unique()
pairs = set(it.product(unique,unique))
aggFlows = pd.DataFrame(list(pairs)).rename(columns={0:'Origin', 1:'Destination'})
aggFlows['Trips'] = 0
aggFlows['Dij'] = 0
```

```python
trips = pd.read_csv('C:\\Users\\Taylor\\Desktop\\barclayscyclehireusagestats2012\\barclayscyclehireusagestats2012\\2012\\someData.csv')
trips = trips[trips['EndStation Id'].isin(list(stations['Field1']))]
trips = trips[trips['StartStation Id'].isin(list(stations['Field1']))]
trips = trips[['StartStation Id','EndStation Id']].groupby(['StartStation Id','EndStation Id']).size().reset_index()
trips = trips[0:10000]
```

```python
def agg_flows(stations, london, aggFlows, x):
    start = stations[stations['Field1'] == x['StartStation Id']]
    end = stations[stations['Field1'] == x['EndStation Id']]
    start = london[london.geometry.intersects(start.geometry.unary_union) == True]['NAME'].values[0]
    end = london[london.geometry.intersects(end.geometry.unary_union) == True]['NAME'].values[0]
    aggFlows['Trips'][((aggFlows['Origin'] == start) & (aggFlows['Destination'] == end))] += x[0]
```

```python
t0 = time.time()
trips.apply(lambda x: agg_flows(stations, london, aggFlows, x), axis=1)
t1 = time.time()
print t1-t0
```

```python
ward_with_stations['centroid'] = ward_with_stations.centroid
```

```python
def distances(ward_with_stations, x):
    origin = x.Origin
    destination = x.Destination
    originGeom = ward_with_stations['centroid'][ward_with_stations['NAME'] == origin].values[0]
    destinationGeom = ward_with_stations['centroid'][ward_with_stations['NAME'] == destination].values[0]
    aggFlows['Dij'][((aggFlows['Origin'] == origin) & (aggFlows['Destination'] == destination))] = originGeom.distance(destinationGeom)
    
```

```python
aggFlows.apply(lambda x: distances(ward_with_stations, x), axis=1)
```
