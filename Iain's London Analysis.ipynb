```python
import pySI as SI
import pandas as pd
import numpy as np
import time


data = pd.DataFrame(pd.read_csv("C:\\Users\\Taylor\\Documents\\london_flows.csv"))
model = SI.calibrate(data=data, trips='Data', sep='distance', cost='invpow', factors={'destinations':['dj']}, constraints={'production':'Origin'})
results = model.mle(initialParams={'beta':0, 'dj':1})

    
#results = model.regression()
```

```python
flows = []
output = {}
data['pred_Flows'] = ''
data['R_Squared'] = ''
data['dj_Param'] = ''
data['decay_Param'] = ''
data['dj_SE'] = ''
data['decay_SE'] = ''
data['dj_T'] = ''
data['decay_T'] = ''
data['dj_P'] = ''
data['decay_P'] = ''
for origin in data.Origin.unique():
    print origin
    newData = data[data.Origin == origin]
    model = SI.calibrate(data=newData, trips='Data', sep='distance', cost='invpow', factors={'destinations':['dj']}, constraints={'production':'Origin'})
    results = model.regression()
    flows.append(results.results.fittedvalues.to_dict().values())
    data['R_Squared'].ix[data['Origin'] == origin] = results.results.rsquared
    data['dj_Param'].ix[data['Origin'] == origin] = results.results.params[1]
    data['decay_Param'].ix[data['Origin'] == origin] = results.results.params[2]
    data['dj_SE'].ix[data['Origin'] == origin] = results.results.bse[1]
    data['decay_SE'].ix[data['Origin'] == origin] = results.results.bse[2]
    data['dj_T'].ix[data['Origin'] == origin] = results.results.tvalues[1]
    data['decay_T'].ix[data['Origin'] == origin] = results.results.tvalues[2]
    data['dj_P'].ix[data['Origin'] == origin] = results.results.pvalues[1]
    data['decay_P'].ix[data['Origin'] == origin] = results.results.pvalues[2]
    
    
    
```
