```python

#set of currently used station's id values 
stationsSet = set(stations[0].values)
    
#set of all station ids for comparison
stationList = set(range(stations[0].values.max()))
    
#stations which there is no coordinates for but exists in the trip data
notUsed = stationList.difference(stationsSet)

#Filter out trips which start or end at a station with no coords
for station in notUsed:
    data = data[(data['EndStation Id'] != station)]
    data = data[(data['StartStation Id'] != station)]



#Get all starting station ids from trip data
startID = set((data["StartStation Id"].values))

#Get all ending station ids from trip data
endID = set((data["EndStation Id"].values))

#stations used for all trips are those where any trip has started or ended
stationsUsed = startID.union(endID)


nodes={}
#list to keep track of stations without proper coords
noStation = []
for station in stationsUsed:
    #if no coords add to list
    if len((stations[[1, 2]][stations[0] == station].values).flatten()) < 2:
        noStation.append(station)
    #otherwise add to node dict
    else:
        lon,lat = (stations[[1, 2]][stations[0] == station].values).flatten()
        nodes[station] = [lon, lat]
        
#count number of trips between unique origin/destination pairs        
data = data[['StartStation Id','EndStation Id']].groupby(['StartStation Id','EndStation Id']).size().reset_index()

#rename column just for ease
data = pd.DataFrame(data.rename(columns = {0:'count'}))

#tuple of edge pairs and weight (trip count)
edges = zip(data['StartStation Id'], data['EndStation Id'], [dict(weight=data['count'].ix[x]) for x in range(len(data['count']))])

#construct graph and plot it
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_weighted_edges_from(edges)
plt.figure(1,figsize=(20,20), dpi=80) 
nx.draw(G, nodes)
plt.draw()

print data





     
```
